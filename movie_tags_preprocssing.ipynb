```python
import pandas as pd
from pathlib import Path

src = Path("./ml-25m/ml-25m/genome-scores.csv")   
dst = src.with_name("genome-scores-0.5.csv")                      

# 결과 파일이 있으면 지우고 시작(덮어쓰기)
if dst.exists():
    dst.unlink()

chunksize = 500_000  # 메모리에 맞춰 조절
use_cols  = ["movieId", "tagId", "relevance"]
dtypes    = {"movieId": "int64", "tagId": "int64", "relevance": "float64"}

written_header = False

for chunk in pd.read_csv(src, usecols=use_cols, dtype=dtypes, chunksize=chunksize):
    # 0.5 이하는 삭제 → 0.5 초과만 남김
    filt = chunk[chunk["relevance"] > 0.5]
    if not filt.empty:
        filt.to_csv(dst, index=False, mode="a", header=not written_header)
        written_header = True

print("완료:", dst, "생성됨")
```

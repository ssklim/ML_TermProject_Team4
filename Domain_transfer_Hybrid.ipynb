```python
import pandas as pd

movie = pd.read_csv('/content/movie_tot.csv')
movie
```

```python
book = pd.read_csv('/content/content/goodbooks.csv')
book
```

```python
import os
import numpy as np
from typing import Optional, List, Tuple

from scipy.sparse import coo_matrix, csr_matrix
from sklearn.decomposition import TruncatedSVD
from sklearn.linear_model import RidgeCV
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import normalize
from sentence_transformers import SentenceTransformer
```

```python
# ---------------- 컬럼 매핑 ----------------
movie_cols = movie.columns.str.lower()
uid_col = "user_id" if "user_id" in movie_cols else [c for c in movie.columns if c.lower().startswith("user")][0]
iid_col = "movie_id" if "movie_id" in movie_cols else [c for c in movie.columns if "movie" in c.lower() and "id" in c.lower()][0]
rat_col = "rating"  if "rating"  in movie_cols else [c for c in movie.columns if "rating" in c.lower()][0]

movie_text_col_candidates = [c for c in movie.columns if c.lower() in ["tag_list","tags","overview","description","genres","content"]]
if not movie_text_col_candidates:
    movie_text_col_candidates = [c for c in movie.columns if c.lower() in ["title","original_title"]]
movie_text_col = movie_text_col_candidates[0]

book_cols = book.columns.str.lower()
bid_col = "book_id" if "book_id" in book_cols else [c for c in book.columns if "book" in c.lower() and "id" in c.lower()][0]
book_text_col_candidates = [c for c in book.columns if c.lower() in ["tag_list","tags","overview","description","categories","content"]]
if not book_text_col_candidates:
    book_text_col_candidates = [c for c in book.columns if c.lower() in ["title"]]
book_text_col = book_text_col_candidates[0]
```

```python
# ---------------- 영화 평점 행렬 ----------------
user_ids = movie[uid_col].astype("int64").values
item_ids = movie[iid_col].astype("int64").values
ratings  = movie[rat_col].astype("float32").values

uniq_users, u_idx = np.unique(user_ids, return_inverse=True)
uniq_items, i_idx = np.unique(item_ids, return_inverse=True)
n_users, n_items  = len(uniq_users), len(uniq_items)

df_tmp = pd.DataFrame({"u":u_idx, "r":ratings})
user_mean = df_tmp.groupby("u")["r"].mean().astype("float32")
centered_r = ratings - user_mean.loc[u_idx].values

R = coo_matrix((centered_r, (u_idx, i_idx)), shape=(n_users, n_items)).tocsr()
```

```python
from sklearn.preprocessing import normalize as _norm
```

```python
# ---------------- SVD 잠재공간 ----------------
rank_k = 64
svd = TruncatedSVD(n_components=rank_k, random_state=42)
movie_user_factors = _norm(svd.fit_transform(R))
movie_item_factors = _norm(svd.components_.T)

user_to_index = {u:i for i,u in enumerate(uniq_users)}
item_to_index = {m:i for i,m in enumerate(uniq_items)}
movie_itemid_array = uniq_items  # 순서 보존
```

```python
# ---------------- 텍스트 임베딩 ----------------
def _prep_text(s: str) -> str:
    return str(s).replace(",", " ").replace("/", " ").replace("|", " ")

sbert_model_name = "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
sbert = SentenceTransformer(sbert_model_name)

movie_items = movie[[iid_col, movie_text_col]].drop_duplicates(subset=[iid_col]).reset_index(drop=True)
movie_items["__text__"] = movie_items[movie_text_col].map(_prep_text)

book_items = book[[bid_col, book_text_col, *(["title"] if "title" in book.columns else [])]].drop_duplicates(subset=[bid_col]).reset_index(drop=True)
book_items["__text__"] = book_items[book_text_col].map(_prep_text)

movie_texts = movie_items["__text__"].tolist()
book_texts  = book_items["__text__"].tolist()

movie_content_emb = sbert.encode(movie_texts, batch_size=256, convert_to_numpy=True, normalize_embeddings=True)
book_content_emb  = sbert.encode(book_texts,  batch_size=256, convert_to_numpy=True, normalize_embeddings=True)
```

```python
# ---------------- 콘텐츠→영화잠재 매핑 ----------------
mid_to_row = {mid:i for i,mid in enumerate(movie_items[iid_col].values)}
X_train, Y_train = [], []
for mid, j in item_to_index.items():
    row = mid_to_row.get(mid, None)
    if row is not None:
        X_train.append(movie_content_emb[row])
        Y_train.append(movie_item_factors[j])

X_train = np.vstack(X_train)
Y_train = np.vstack(Y_train)

use_mlp = False
if not use_mlp:
    reg = RidgeCV(alphas=(1e-3,1e-2,1e-1,1,10))
else:
    reg = MLPRegressor(hidden_layer_sizes=(512,256), activation="relu",
                       solver="adam", random_state=42, max_iter=50, verbose=False)
reg.fit(X_train, Y_train)

book_latent = _norm(reg.predict(book_content_emb))
bookid_array = book_items[bid_col].values
bookid_to_index = {b:i for i,b in enumerate(bookid_array)}
```

```python
from numpy.linalg import lstsq
```

```python
# ---------------- 공통 유틸 ----------------
def _fit_affine(x: np.ndarray, y: np.ndarray) -> Tuple[float, float]:
    x = np.asarray(x).reshape(-1,1)
    y = np.asarray(y).reshape(-1,1)
    if len(x) < 2 or np.isclose(x.std(), 0):
        a = 1.0
        b = float(y.mean() - x.mean())
        return a, b
    X = np.hstack([x, np.ones_like(x)])
    a, b = lstsq(X, y, rcond=None)[0].ravel()
    return float(a), float(b)

def rmse(y_true: np.ndarray, y_pred: np.ndarray) -> float:
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    return float(np.sqrt(np.mean((y_true - y_pred) ** 2)))

def recall_at_k(recommended_ids: np.ndarray, true_ids: set, k: int) -> float:
    rec_k = recommended_ids[:k]
    hit = sum(1 for x in rec_k if x in true_ids)
    return hit / max(1, len(true_ids))

def ndcg_at_k(recommended_ids: np.ndarray, true_ids: set, k: int) -> float:
    dcg = 0.0
    for i, iid in enumerate(recommended_ids[:k], start=1):
        if iid in true_ids:
            dcg += 1.0 / np.log2(i + 1)
    ideal_hits = min(k, len(true_ids))
    idcg = sum(1.0 / np.log2(i + 1) for i in range(1, ideal_hits + 1)) if ideal_hits > 0 else 0.0
    return dcg / idcg if idcg > 0 else 0.0
```

```python
# ---------------- 책 영역 ----------------
b_uid = "user_id" if "user_id" in {c.lower():c for c in book.columns} else [c for c in book.columns if "user" in c.lower()][0]
b_iid = bid_col
b_rat = "rating" if "rating" in {c.lower():c for c in book.columns} else None
book_inter = book[[b_uid, b_iid] + ([b_rat] if b_rat else [])].dropna(subset=[b_iid]).copy()

def _book_ids_to_latents(ids: np.ndarray) -> np.ndarray:
    idx = [bookid_to_index[i] for i in ids if i in bookid_to_index]
    if not idx: return np.zeros((0, book_latent.shape[1]), dtype=np.float32)
    return book_latent[idx]

def _weights_from_ratings(r):
    if r is None or len(r)==0: return None
    r = r.astype(float)
    if r.max() == r.min(): return np.ones_like(r) / len(r)
    w = (r - r.min()) / (r.max() - r.min() + 1e-6)
    w = w + 1e-3
    return w / w.sum()

def _user_vec_books(book_ids, ratings):
    Z = _book_ids_to_latents(book_ids)
    if Z.shape[0]==0: return np.zeros((book_latent.shape[1],), dtype=np.float32)
    if ratings is not None:
        w = _weights_from_ratings(ratings)
        u = (Z * w[:,None]).sum(axis=0)
    else:
        u = Z.mean(axis=0)
    return _norm(u.reshape(1,-1)).ravel()

def recommend_books_for_book_user(book_user_id: int, top_k: int = 10, holdout: int = 0) -> pd.DataFrame:
    dfu = book_inter[book_inter[b_uid]==book_user_id].sort_index()
    if dfu.empty: raise ValueError("해당 책 사용자 상호작용이 없음.")
    train = dfu.iloc[:-holdout] if holdout>0 and len(dfu)>holdout else dfu
    u_vec = _user_vec_books(
        train[b_iid].values,
        train[b_rat].values if b_rat and b_rat in train.columns else None
    )
    if not np.any(u_vec): raise ValueError("사용자 벡터 생성 실패.")
    seen = set(train[b_iid].values.tolist())
    scores_all = book_latent @ u_vec
    cand_idx = np.arange(len(book_items))
    if seen:
        mask = np.array([bookid_array[i] not in seen for i in range(len(book_items))])
        cand_idx = cand_idx[mask]
        scores = scores_all[mask]
    else:
        scores = scores_all
    top_idx = np.argpartition(scores, -top_k)[-top_k:]
    top_idx = top_idx[np.argsort(scores[top_idx])[::-1]]
    out = book_items.iloc[cand_idx[top_idx], :].copy()
    out["score"] = scores[top_idx]
    cols = [c for c in ["title", b_iid, "score", book_text_col] if c in out.columns]
    return out[cols].reset_index(drop=True)

def evaluate_on_books(holdout: int = 1, k_list = [5,10], max_users: int = 1000) -> pd.DataFrame:
    if not b_rat or b_rat not in book_inter.columns:
        raise ValueError("book에 rating이 없어 RMSE 불가.")
    results = []
    users = book_inter[b_uid].value_counts().index.tolist()[:max_users]
    for u in users:
        dfu = book_inter[book_inter[b_uid]==u].sort_index()
        if len(dfu)<=holdout: continue
        train, test = dfu.iloc[:-holdout], dfu.iloc[-holdout:]
        u_vec = _user_vec_books(
            train[b_iid].values,
            train[b_rat].values if b_rat and b_rat in train.columns else None
        )
        if not np.any(u_vec): continue
        seen = set(train[b_iid].values.tolist())
        scores_all = book_latent @ u_vec
        cand_idx = np.arange(len(book_items))
        if seen:
            mask = np.array([bookid_array[i] not in seen for i in range(len(book_items))])
            cand_idx = cand_idx[mask]
            scores = scores_all[mask]
        else:
            scores = scores_all
        order = np.argsort(-scores)
        rec_ids = bookid_array[cand_idx[order]]
        true_ids = set(test[b_iid].values.tolist())

        row = {"user": u}
        for k in k_list:
            row[f"Recall@{k}"] = recall_at_k(rec_ids, true_ids, k)
            row[f"NDCG@{k}"]   = ndcg_at_k(rec_ids, true_ids, k)

        train_lat = _book_ids_to_latents(train[b_iid].values)
        if train_lat.shape[0] >= 1:
            train_scores  = (train_lat @ u_vec).ravel()
            train_ratings = train[b_rat].astype(float).values.ravel()
            a,b = _fit_affine(train_scores, train_ratings)
            test_lat  = _book_ids_to_latents(test[b_iid].values)
            test_scr  = (test_lat @ u_vec).ravel()
            y_pred = a*test_scr + b
            y_true = test[b_rat].astype(float).values.ravel()
            rmin, rmax = float(book_inter[b_rat].min()), float(book_inter[b_rat].max())
            y_pred = np.clip(y_pred, rmin, rmax)
            row["RMSE"] = rmse(y_true, y_pred)
        else:
            row["RMSE"] = np.nan

        results.append(row)
    return pd.DataFrame(results)
```

```python
# ---------------- 영화 영역 ----------------
def _movie_ids_to_latents(ids: np.ndarray) -> np.ndarray:
    idx = [item_to_index[i] for i in ids if i in item_to_index]
    if not idx: return np.zeros((0, movie_item_factors.shape[1]), dtype=np.float32)
    return movie_item_factors[idx]

def _user_vec_movies(movie_ids, ratings):
    Z = _movie_ids_to_latents(movie_ids)
    if Z.shape[0]==0: return np.zeros((movie_item_factors.shape[1],), dtype=np.float32)
    if ratings is not None:
        r = ratings.astype(float)
        r = r - r.mean()
        if np.allclose(r.max(), r.min()):
            w = np.ones_like(r)/len(r)
        else:
            w = (r - r.min())/(r.max()-r.min()+1e-6); w = w + 1e-3; w = w/w.sum()
        u = (Z * w[:,None]).sum(axis=0)
    else:
        u = Z.mean(axis=0)
    return _norm(u.reshape(1,-1)).ravel()

def recommend_movies_for_user(movie_user_id: int, top_k: int = 10, holdout: int = 0) -> pd.DataFrame:
    dfu = movie[movie[uid_col]==movie_user_id].sort_index()
    if dfu.empty: raise ValueError("해당 영화 사용자 상호작용이 없음.")
    train = dfu.iloc[:-holdout] if holdout>0 and len(dfu)>holdout else dfu
    u_vec = _user_vec_movies(
        train[iid_col].astype(int).values,
        train[rat_col].astype(float).values if rat_col in train.columns else None
    )
    if not np.any(u_vec): raise ValueError("사용자 벡터 생성 실패.")
    seen = set(train[iid_col].astype(int).values.tolist())
    scores_all = movie_item_factors @ u_vec
    cand_idx = np.arange(len(movie_itemid_array))
    if seen:
        mask = np.array([movie_itemid_array[i] not in seen for i in range(len(movie_itemid_array))])
        cand_idx = cand_idx[mask]
        scores = scores_all[mask]
    else:
        scores = scores_all
    top_idx = np.argpartition(scores, -top_k)[-top_k:]
    top_idx = top_idx[np.argsort(scores[top_idx])[::-1]]

    mv_meta = movie[[iid_col, movie_text_col]].drop_duplicates(subset=[iid_col])
    out_ids = movie_itemid_array[cand_idx[top_idx]]
    out = mv_meta[mv_meta[iid_col].isin(out_ids)].copy()
    order_map = {mid:i for i,mid in enumerate(out_ids.tolist())}
    out["__order__"] = out[iid_col].map(order_map)
    out = out.sort_values("__order__").drop(columns="__order__")
    out["score"] = scores[top_idx]
    cols = [c for c in [iid_col, "score", movie_text_col] if c in out.columns]
    return out[cols].reset_index(drop=True)

def evaluate_on_movies(holdout: int = 1, k_list = [5,10], max_users: int = 1000) -> pd.DataFrame:
    results = []
    users = movie[uid_col].value_counts().index.tolist()[:max_users]
    for u in users:
        dfu = movie[movie[uid_col]==u].sort_index()
        if len(dfu)<=holdout: continue
        train, test = dfu.iloc[:-holdout], dfu.iloc[-holdout:]
        u_vec = _user_vec_movies(
            train[iid_col].astype(int).values,
            train[rat_col].astype(float).values if rat_col in train.columns else None
        )
        if not np.any(u_vec): continue
        seen = set(train[iid_col].astype(int).values.tolist())
        scores_all = movie_item_factors @ u_vec
        cand_idx = np.arange(len(movie_itemid_array))
        if seen:
            mask = np.array([movie_itemid_array[i] not in seen for i in range(len(movie_itemid_array))])
            cand_idx = cand_idx[mask]
            scores = scores_all[mask]
        else:
            scores = scores_all
        order = np.argsort(-scores)
        rec_ids = movie_itemid_array[cand_idx[order]]
        true_ids = set(test[iid_col].astype(int).values.tolist())

        row = {"user": int(u)}
        for k in k_list:
            row[f"Recall@{k}"] = recall_at_k(rec_ids, true_ids, k)
            row[f"NDCG@{k}"]   = ndcg_at_k(rec_ids, true_ids, k)

        if rat_col in train.columns and rat_col in test.columns:
            train_lat = _movie_ids_to_latents(train[iid_col].astype(int).values)
            if train_lat.shape[0] >= 1:
                train_scores  = (train_lat @ u_vec).ravel()
                train_ratings = train[rat_col].astype(float).values.ravel()
                a,b = _fit_affine(train_scores, train_ratings)

                test_lat = _movie_ids_to_latents(test[iid_col].astype(int).values)
                test_scr = (test_lat @ u_vec).ravel()
                y_pred = a*test_scr + b
                y_true = test[rat_col].astype(float).values.ravel()

                rmin, rmax = float(movie[rat_col].min()), float(movie[rat_col].max())
                y_pred = np.clip(y_pred, rmin, rmax)
                row["RMSE"] = rmse(y_true, y_pred)
            else:
                row["RMSE"] = np.nan
        else:
            row["RMSE"] = np.nan

        results.append(row)
    return pd.DataFrame(results)

def similar_books(book_id: int, top_k: int = 10) -> pd.DataFrame:
    if book_id not in bookid_to_index: raise ValueError("book_id 미존재")
    i = bookid_to_index[book_id]
    z = book_latent[i]
    sims = book_latent @ z
    sims[i] = -np.inf
    top = np.argpartition(sims, -top_k)[-top_k:]
    top = top[np.argsort(sims[top])[::-1]]
    out = book_items.iloc[top].copy()
    out["sim"] = sims[top]
    cols = [c for c in ["title", bid_col, "sim", book_text_col] if c in out.columns]
    return out[cols].reset_index(drop=True)

def similar_movies(movie_id: int, top_k: int = 10) -> pd.DataFrame:
    if movie_id not in item_to_index: raise ValueError("movie_id 미존재")
    j = item_to_index[movie_id]
    z = movie_item_factors[j]
    sims = movie_item_factors @ z
    sims[j] = -np.inf
    top = np.argpartition(sims, -top_k)[-top_k:]
    top = top[np.argsort(sims[top])[::-1]]
    mv_meta = movie[[iid_col, movie_text_col]].drop_duplicates(subset=[iid_col])
    out_ids = movie_itemid_array[top]
    out = mv_meta[mv_meta[iid_col].isin(out_ids)].copy()
    order_map = {mid:i for i, mid in enumerate(out_ids.tolist())}
    out["__order__"] = out[iid_col].map(order_map)
    out = out.sort_values("__order__").drop(columns="__order__")
    out["sim"] = sims[top]
    cols = [c for c in [iid_col, "sim", movie_text_col] if c in out.columns]
    return out[cols].reset_index(drop=True)
```

```python
# ================== RMSE/MAE 전수평가 전용 블록 ==================
def mae(y_true: np.ndarray, y_pred: np.ndarray) -> float:
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    return float(np.mean(np.abs(y_true - y_pred)))

def _predict_with_affine(u_vec, ids, ratings, ids_to_latent_fn, all_ratings_min, all_ratings_max):
    """train set으로 affine(a,b) 맞춰 test 점수 예측을 평점 스케일로 보정"""
    train_lat = ids_to_latent_fn(ids)
    if train_lat.shape[0] < 1:
        return None  # 사용자 벡터 부족
    train_scores  = (train_lat @ u_vec).ravel()
    train_ratings = ratings.astype(float).values.ravel()
    a, b = _fit_affine(train_scores, train_ratings)
    test_lat = ids_to_latent_fn  # 클로저 활용 위해 반환
    return a, b, test_lat

# ---------------- 책: RMSE/MAE ----------------
def evaluate_rmse_mae_books(holdout: int = 1, max_users: Optional[int] = None) -> pd.DataFrame:
    if not b_rat or b_rat not in book_inter.columns:
        raise ValueError("book 데이터에 rating 컬럼이 없어 RMSE/MAE 계산 불가.")
    users = book_inter[b_uid].value_counts().index.tolist()
    if max_users: users = users[:max_users]

    rmin, rmax = float(book_inter[b_rat].min()), float(book_inter[b_rat].max())
    rows = []
    for u in users:
        dfu = book_inter[book_inter[b_uid] == u].sort_index()
        if len(dfu) <= holdout: continue
        train, test = dfu.iloc[:-holdout], dfu.iloc[-holdout:]

        u_vec = _user_vec_books(
            train[b_iid].values,
            train[b_rat].values if b_rat and b_rat in train.columns else None
        )
        if not np.any(u_vec): continue

        info = _predict_with_affine(
            u_vec, train[b_iid].values, train[b_rat],
            ids_to_latent_fn=_book_ids_to_latents,
            all_ratings_min=rmin, all_ratings_max=rmax
        )
        if info is None: continue
        a, b, _ = info

        test_lat = _book_ids_to_latents(test[b_iid].values)
        if test_lat.shape[0] < 1: continue

        test_scr = (test_lat @ u_vec).ravel()
        y_pred = a * test_scr + b
        y_true = test[b_rat].astype(float).values.ravel()
        y_pred = np.clip(y_pred, rmin, rmax)

        rows.append({"user": int(u), "RMSE": rmse(y_true, y_pred), "MAE": mae(y_true, y_pred)})

    return pd.DataFrame(rows)

# ---------------- 영화: RMSE/MAE ----------------
def evaluate_rmse_mae_movies(holdout: int = 1, max_users: Optional[int] = None) -> pd.DataFrame:
    if rat_col not in movie.columns:
        raise ValueError("movie 데이터에 rating 컬럼이 없어 RMSE/MAE 계산 불가.")
    users = movie[uid_col].value_counts().index.tolist()
    if max_users: users = users[:max_users]

    rmin, rmax = float(movie[rat_col].min()), float(movie[rat_col].max())
    rows = []
    for u in users:
        dfu = movie[movie[uid_col] == u].sort_index()
        if len(dfu) <= holdout: continue
        train, test = dfu.iloc[:-holdout], dfu.iloc[-holdout:]

        u_vec = _user_vec_movies(
            train[iid_col].astype(int).values,
            train[rat_col].astype(float).values if rat_col in train.columns else None
        )
        if not np.any(u_vec): continue

        train_lat = _movie_ids_to_latents(train[iid_col].astype(int).values)
        if train_lat.shape[0] < 1: continue

        train_scores  = (train_lat @ u_vec).ravel()
        train_ratings = train[rat_col].astype(float).values.ravel()
        a, b = _fit_affine(train_scores, train_ratings)

        test_lat = _movie_ids_to_latents(test[iid_col].astype(int).values)
        if test_lat.shape[0] < 1: continue

        test_scr = (test_lat @ u_vec).ravel()
        y_pred = a * test_scr + b
        y_true = test[rat_col].astype(float).values.ravel()
        y_pred = np.clip(y_pred, rmin, rmax)

        rows.append({"user": int(u), "RMSE": rmse(y_true, y_pred), "MAE": mae(y_true, y_pred)})

    return pd.DataFrame(rows)

# ================== 실행: 기술통계만 출력 ==================
books_metrics  = evaluate_rmse_mae_books(holdout=1, max_users=10000)
movies_metrics = evaluate_rmse_mae_movies(holdout=1, max_users=10000)

print("=== Books: RMSE/MAE describe ===")
display(books_metrics[["RMSE","MAE"]].describe().T)

print("=== Movies: RMSE/MAE describe ===")
display(movies_metrics[["RMSE","MAE"]].describe().T)
```

```python
# --- 한 사용자: 높게 평가한 책 Top-20 + 추천 Top-10 -------------------------

def top_rated_books_for_user(book_user_id: int, top_n: int = 20) -> pd.DataFrame:
    """해당 사용자가 높게 평가한 책 상위 N개 반환"""
    if not b_rat or b_rat not in book_inter.columns:
        raise ValueError("book 데이터에 rating 컬럼이 없어 상위 평점 목록을 만들 수 없음.")
    dfu = book_inter[book_inter[b_uid] == book_user_id].copy()
    if dfu.empty:
        raise ValueError("해당 사용자 기록이 없음.")
    # 평점 내림차순, 동일 평점이면 최근(인덱스 큰 것) 우선
    dfu = dfu.sort_values([b_rat], ascending=[False]).head(top_n)
    # 메타 병합(제목 등)
    meta = book_items[[bid_col] + ([ "title"] if "title" in book_items.columns else []) + ([book_text_col] if book_text_col in book_items.columns else [])]
    out = dfu.merge(meta, left_on=b_iid, right_on=bid_col, how="left")
    cols = [c for c in ["title", bid_col, b_rat, book_text_col] if c in out.columns]
    return out[cols].reset_index(drop=True)

# 타깃 사용자 선택: 이전에 만든 ex_book_user가 있으면 사용, 없으면 임의 1명
target_user = ex_book_user if "ex_book_user" in globals() else int(book_inter[b_uid].iloc[0])

print(f"[사용자 {target_user}] 상위 평점 Top-20")
display(top_rated_books_for_user(target_user, top_n=20))

print(f"[사용자 {target_user}] 추천 Top-10 (보지 않은 책만)")
display(recommend_books_for_book_user(target_user, top_k=10, holdout=1))  # holdout=1은 마지막 1권을 테스트로 제외
```
